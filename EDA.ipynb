```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

```python
# Select five countries from the dataset
countries = ['Central African Republic', 'Albania', 'Uzbekistan', 'Morocco', 'Thailand']  
```

```python
# Filter the dataset to include only the selected countries
df_selected = df[df['Country'].isin(countries)]
```

```python
# Pivot the DataFrame to have years as columns and countries as rows
df_pivot = df_selected.pivot(index='Country', columns='Year', values='Democracy Score')
```

```python
# Plotting the line graph
ax = df_pivot.T.plot(kind='line')

# Customize the legend
ax.legend(title='Country', loc='center left', bbox_to_anchor=(1, 0.75), fontsize='small')

# Customize the graph
plt.title('Democracy Scores Over Times')
plt.xlabel('Year')
plt.ylabel('Democracy Score')

# Display the graph
plt.show()
```
